A distribuição cumulativa empírica é uma função que explica como valores associadas a uma probabilidade. Dados os exemplos, o primeiro gráfico, temos que SepalLength (Comprimento da sépala) tem 40% dos dados tem 5,5 cm. Essa ferramenta traz um panorama empírico da característica.

```python
# Empirical Cumulative Distribution
# visualising ECDF

from mlxtend.plotting import ecdf
import matplotlib.pyplot as plt

# Criando uma figura e eixos para os subplots
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(13, 40))

# Ajustando o espaçamento entre os subplots
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

# Limpando a figura
plt.clf()

# Iterando sobre cada coluna do dataset
for i, feature in enumerate(dataset.columns, 1):
    # Adicionando um subplot para cada coluna
    plt.subplot(len(dataset.columns), 3, i)

    # Criando um gráfico de funções de distribuição cumulativa
    ecdf(dataset[feature])

    # Adicionando título ao subplot
    plt.title(f'{feature}', size=15, fontsize=12, fontweight='medium')

    # Removendo as grades do gráfico
    plt.grid(False)

    # Ajustando o tamanho das etiquetas nos eixos x e y
    plt.tick_params(axis='x', labelsize=8)
    plt.tick_params(axis='y', labelsize=8)

# Ajustando o layout dos subplots
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
